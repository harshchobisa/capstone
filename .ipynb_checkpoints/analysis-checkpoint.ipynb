```python
import pandas as pd
import numpy as np
import pickle
```

```python
with open('cifar10_sorted_fulldata.pkl', 'rb') as f:
    forget = pickle.load(f)
```

```python
from numpy import load

full_craig = load('craig_cifar10.npz')
lst = full_craig.files
for item in lst:
    print(item)
    print(full_craig[item])
```

```python
craig = full_craig['subset'][0]
```

```python
craig.shape
```

```python
np.std(forget['forgetting counts'].tolist())
```

```python
forget
```

```python
forget_counts_dict = {}
for i, val in enumerate(forget['indices']):
    forget_counts_dict[val] = forget['forgetting counts'][i]
```

```python
forget_counts_dict
```

```python
# not_selected_sum = np.sum(full_craig['not_selected'][0])
# print(full_craig['not_selected'].shape)

not_selected_sum = []

for i in full_craig['not_selected'][0]:
    not_selected_sum.append(forget_counts_dict[int(i)])
    
np.mean(not_selected_sum)
np.std(not_selected_sum)
```

```python
all_sums = []
for i in range(200):
    sums = 0
    for j in range(5000):
        sums += forget_counts_dict[int(craig[i][j])]
    all_sums.append(sums)
        
```

```python
average_forgettability = np.array(all_sums) / 5000
```

```python
import matplotlib.pyplot as plt

plt.plot(range(200), all_sums)
plt.xlabel("epoch")
plt.ylabel("forgettabillity total")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(range(200), average_forgettability)
plt.xlabel("epoch")
plt.ylabel("forgettabillity total")
plt.show()
```
